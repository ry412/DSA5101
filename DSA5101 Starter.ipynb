# Logistic Regression

### Spliting the training set
We will split our training set here to train/validation sets for our purpose of feature selection, as we do not wish to expose any information about the test set before model evaluation.

```python
train, validation = train_test_split(data, test_size=0.2, random_state=7)
```

```python
print(f'train.shape: {train.shape}')
print(f'validation.shape: {validation.shape}')
```

```python
# set our target variable
target = 'y'
```

### Overview of the quality of our data set

Here, we caluculate `iv`, `gini`, `entropy` to get a general idea of the quality of our data(in term of features), where IV (information value) is calculated as:
$IV = \sum(Event\% - Non event\%) * \ln((Event\%) \div (Non event\%))$ <br>
Similar to gini and entropy, IV reflects how well this feature can classify our target variable. The larger the IV, the stronger of a classifier the feature is.

```python
quality = toad.quality(train, target = 'y')
quality
```

```python
# Here, we eliminate the features who have iv<=0.02 as they might be too weak of predictors.
selected_features = list(quality[quality['iv']>0.02].index)
selected_features.append(target)
len(selected_features)
```

### Feature binning and WOE transformation

Here, we perform feature binning, which in other words, data discretization. <br>
By doing this, it enables us to perform WOE(Weight of Evidence) transformation on the data, which handles missing values/outliers and scales/encodes our data. Moreover, WOE transformation could establish a monotonic relationship between our features and target variable.<br>
$WOE = \ln((Event\%) \div (Non event\%))$

```python
c = toad.transform.Combiner()
c.fit(train, y = target, method = 'dt',min_samples = 0.03, n_bins = 10)#n_bins = 5,min_samples = 0.05
#empty_separate = False
train0 = c.transform(train)
test0 = c.transform(test)
validation0 = c.transform(validation)
```

```python
# WOE transformation
t = toad.transform.WOETransformer()
train_woe = t.fit_transform(train0,train0[target], exclude=target)
test_woe = t.transform(test0)
validation_woe = t.transform(validation0)
```

### Feature correlation 

In this section, as we do not have too many features, we will draw a correlation heatmap to help us understanding the correlations between features, and we will perform feature selection if there is any feature that have high correlation with others.

```python
corr_df = train.corr()
```

```python
plt.figure(figsize=(25, 9))
sns.heatmap(corr_df,annot=True ,cmap=sns.color_palette("BrBG",2));
```

As we do not observe any feature that has high correlation with others, we won't drop any features at this step.

### binning plot

We will now plot the binning figure along with the event rate for each feature. Since we do not want to use the information of the test set except for model evaluation, we will plot the figures using our training and validation set. We will also perfrom manual feature selection accordingly when we observe inconsistency of event rates between the train and validation set, low differenct of the event rate between bins etc.

```python
book = xlsxwriter.Workbook('./pict.xlsx')
sheet = book.add_worksheet('binning')
path = './binplot'
num1 = 1
ima1 = 1
try:
    os.mkdir(path)
except OSError as error:
    print(error)
    
for names in train.drop(target,axis=1).columns:
    fig1 = bin_plot(train0,x=names,target=target).get_figure()
    fig2 = bin_plot(validation0,x=names,target=target).get_figure()
    imadir1 = path + '/image'+ str(ima1) + r'.png' 
    imadir2 = path + '/image'+ str(ima1 +1) + r'.png' 
    fig1.savefig(imadir1)
    fig2.savefig(imadir2)
    wordic1 = 'A' + str(num1)
    plotdic1 = 'A' + str(num1 + 1)
    plotdic2 = 'J' + str(num1 + 1)
    sheet.write(wordic1,names)
    sheet.insert_image(plotdic1,imadir1,{'x_scale': 0.5,'y_scale':0.65})
    sheet.insert_image(plotdic2,imadir2,{'x_scale': 0.5,'y_scale':0.65})
    num1 = num1 + 20
    ima1 = ima1 + 2
book.close()
```

From observing the graph above, we don't observer any significant inconsistency and weak predicators. Thus, we will keep all of our features to achieve the best performance possible. However, as there is a trade-off between feature numbers and robustness of the model, we should adjust the feature selection strategies here according to what we want to achieve.

### Modeling

```python
train_woe = train_woe[selected_features]
test_woe = test_woe[selected_features]
validation_woe = validation_woe[selected_features]
```

```python
model = LogisticRegression(C=1, solver='newton-cg')
model.fit(train_woe.drop(target,axis=1),train_woe[target])
```

```python
y_train = model.predict_proba(train_woe.drop(target,axis=1))[:,1]
y_test = model.predict_proba(test_woe.drop(target,axis=1))[:,1]
```

# Model Evaluation

```python
y_train_pred = model.predict(train_woe.drop(target,axis=1))
y_test_pred = model.predict(test_woe.drop(target,axis=1))
y_train_true = train[target]
y_test_true = test[target]
```

```python
y_train_pred_prob = model.predict_proba(train_woe.drop(target,axis=1))[:,1]
y_test_pred_prob = model.predict_proba(test_woe.drop(target,axis=1))[:,1]
```

### Training set

```python
print(f'Accuracy for the training set: {accuracy_score(y_train_true, y_train_pred)}')
print(f'Recall for the training set: {recall_score(y_train_true, y_train_pred)}')
print(f'Precision for the training set: {precision_score(y_train_true, y_train_pred)}')
print(f'F-1 for the training set: {f1_score(y_train_true, y_train_pred)}')
```

### Test set

```python
print(f'Accuracy for the test set: {accuracy_score(y_test_true, y_test_pred)}')
print(f'Recall for the test set: {recall_score(y_test_true, y_test_pred)}')
print(f'Precision for the test set: {precision_score(y_test_true, y_test_pred)}')
print(f'F-1 for the test set: {f1_score(y_test_true, y_test_pred)}')
```

### ROC curve

```python
print(f'AUC for Logistic Regression is: {roc_auc_score(y_test_true, y_test_pred_prob)}')
```

```python
false_positive_rate, true_positive_rate, threshold = roc_curve(y_test_true, y_test_pred_prob)
plt.subplots(1, figsize=(10,10))
plt.title('Receiver Operating Characteristic - Logistic Regression')
plt.plot(false_positive_rate, true_positive_rate)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
