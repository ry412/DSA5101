## 1.Logistic Regression

### 1.1 Spliting the training set
We will split our training set here to train/validation sets for our purpose of feature selection, as we do not wish to expose any information about the test set before model evaluation.

```python
train_lr, validation_lr = train_test_split(data, test_size=0.2, random_state=7)
```

```python
print(f'train_lr.shape: {train_lr.shape}')
print(f'validation_lr.shape: {validation_lr.shape}')
```

```python
# set our target variable
target = 'y'
```

### 1.2 Overview of the quality of our data set

Here, we caluculate `iv`, `gini`, `entropy` to get a general idea of the quality of our data(in term of features), where IV (information value) is calculated as:
$$IV = \sum(Event\% - Non event\%) * \ln((Event\%) \div (Non event\%))$$
Similar to gini and entropy, IV reflects how well this feature can classify our target variable. The larger the IV, the stronger of a classifier the feature is.

```python
quality = toad.quality(train_lr, target = 'y')
quality
```

```python
# Here, we eliminate the features who have iv<=0.02 as they might be too weak of predictors.
selected_features = list(quality[quality['iv']>0.02].index)
selected_features.append(target)
len(selected_features)
```

### 1.3 Feature binning and WOE transformation

Here, we perform feature binning, which in other words, data discretization. <br>
By doing this, it enables us to perform WOE(Weight of Evidence) transformation on the data, which handles missing values/outliers and scales/encodes our data. Moreover, WOE transformation could establish a monotonic relationship between our features and target variable.
$$WOE = \ln((Event\%) \div (Non event\%))$$

```python
c = toad.transform.Combiner()
c.fit(train_lr, y = target, method = 'dt',min_samples = 0.03, n_bins = 15)# decision tree
# c.fit(train_lr, y = target, method = 'chi',min_samples = 0.05, n_bins = 5) # chi-squared binning
train_lr0 = c.transform(train_lr)
test_lr0 = c.transform(test)
validation_lr0 = c.transform(validation_lr)
```

```python
# WOE transformation
t = toad.transform.WOETransformer()
train_lr_woe = t.fit_transform(train_lr0,train_lr0[target], exclude=target)
test_lr_woe = t.transform(test_lr0)
validation_lr_woe = t.transform(validation_lr0)
```

### 1.4 binning plot

We will now plot the binning figure along with the event rate for each feature. Since we do not want to use the information of the test set except for model evaluation, we will plot the figures using our train_lring and validation_lr set. We will also perfrom manual feature selection accordingly when we observe inconsistency of event rates between the train_lr and validation_lr set, low differenct of the event rate between bins etc.

```python
for names in train_lr.drop(target,axis=1).columns:
    fig1 = bin_plot(train_lr0,x=names,target=target).get_figure()
    fig2 = bin_plot(validation_lr0,x=names,target=target).get_figure()
```

From observing the graph above, we don't observer any significant inconsistency and weak predicators. Thus, we will keep all of our features to achieve the best performance possible. However, as there is a trade-off between feature numbers and robustness of the model, we should adjust the feature selection strategies here according to what we want to achieve.

### 1.5 Modeling

```python
train_lr_woe = train_lr_woe[selected_features]
test_lr_woe = test_lr_woe[selected_features]
validation_lr_woe = validation_lr_woe[selected_features]
```

#### 1.6.3 Threshold and confusion matrices

As we don't know the business strategy of this campaign, we can not tell whether we have the best threshold for the LR classifier. However, below we compute the `accuracy`, `recall`, `precision`, `f1` as well as the `confusion matrix` for each threshold with a step size of 0.05. This could help us to choose the best threshold we want to choose in real life senarios.

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix

tt = pd.DataFrame(y_test_lr_true)

pred_proba_df = pd.DataFrame(model.predict_proba(test_lr_woe.drop(target,axis=1)))
threshold_list = [0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,.7,.75,.8,.85,.9,.95,.99]
for i in threshold_list:
    print ('\n******** For i = {} ******'.format(i))
    yy_ttest_lr_pred = pred_proba_df.applymap(lambda x: 1 if x>i else 0)
    test_lr_accuracy = metrics.accuracy_score(tt.values.reshape(tt.values.size,1),
                                           yy_ttest_lr_pred.iloc[:,1].values.reshape(yy_ttest_lr_pred.iloc[:,1].values.size,1))
    test_lr_recall = metrics.recall_score(tt.values.reshape(tt.values.size,1),
                                           yy_ttest_lr_pred.iloc[:,1].values.reshape(yy_ttest_lr_pred.iloc[:,1].values.size,1))
    test_lr_precision = metrics.precision_score(tt.values.reshape(tt.values.size,1),
                                           yy_ttest_lr_pred.iloc[:,1].values.reshape(yy_ttest_lr_pred.iloc[:,1].values.size,1))
    test_lr_f1 = metrics.f1_score(tt.values.reshape(tt.values.size,1),
                                           yy_ttest_lr_pred.iloc[:,1].values.reshape(yy_ttest_lr_pred.iloc[:,1].values.size,1))
    print('Our testing accuracy is {}'.format(test_lr_accuracy))
    print('Our testing recall is {}'.format(test_lr_recall))
    print('Our testing precision is {}'.format(test_lr_precision))
    print('Our testing F1 is {}'.format(test_lr_f1))
    print(confusion_matrix(tt.values.reshape(tt.values.size,1),
                           yy_ttest_lr_pred.iloc[:,1].values.reshape(yy_ttest_lr_pred.iloc[:,1].values.size,1)))
```

## 2. SVC

## 2.1 Feature Engineering

As we can see from the EDA part, the original data have some missing value and outliers. And for SVM, it is more sensitive to missing data and noisy data. So before we start model training, we will begin with feature engineering to make sure that data is appropriate for our SVM model.

### 2.1.1 Handling Missing Values      
Here, we drop feature poutcome and pdays as it has over 70% meaningless value.

```python
data0 = data.drop(columns='poutcome').drop(columns='pdays')
test0 = test.drop(columns='poutcome').drop(columns='pdays')
```

```python
data0.shape
```

For job column has little unknown value, we delete the row containing 'unknown'.

```python
data1 = data0[~data0['job'].isin(["unknown"])]
test1 = test0[~test0['job'].isin(["unknown"])]
```

```python
data1.shape
```

For unknown values in conlumn educaton and contact, we are going to replace them with plural.

```python
data2 = data1.copy()
data2['education'].replace('unknown', data2["education"].value_counts().index[0],inplace=True)
data2['contact'].replace('unknown', data2["contact"].value_counts().index[0],inplace=True)
test2 = test1.copy()
test2['education'].replace('unknown', test2["education"].value_counts().index[0],inplace=True)
test2['contact'].replace('unknown', test2["contact"].value_counts().index[0],inplace=True)
```

```python
test2.head()
```

### 2.1.2 Feature Encoding

```python
encoder = ce.OrdinalEncoder(cols=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month'])
data3 = encoder.fit_transform(data2)
test3 = encoder.transform(test2)
```

```python
data3.head()
```

### 2.1.3 Feature Normalization

```python
col = ['age','balance','day','duration','campaign']
```

```python
data4 = data3.copy()
test4 = test3.copy()
```

```python
zscore = preprocessing.StandardScaler()
zscore = zscore.fit(data4.loc[:,col])
data4.loc[:,col] = zscore.transform(data4.loc[:,col])
test4.loc[:,col] = zscore.transform(test4.loc[:,col])
```

```python
data4.head()
```

### 2.1.4 Feature Selection

Here, we first caluculate IV of every features.  As we know that in binary classification problem, the IV value is used to measure the importance of a feature, i.e., how much information this feature can provide in order to predict the task.

```python
quality = toad.quality(data4, target = 'y')
quality
```

Then we conduct preliminary feature selection according to IV and correlation (with other features), the variables are:
1.iv=0.02: the features with IV smaller than 0.02 are eliminated;
2.corr=0.7: if two or more features have Pearson correlation larger than 0.7, the ones with lower IV are eliminated;

```python
data_first_selected, dropped = toad.selection.select(data4,target = 'y', iv = 0.02, corr = 0.7, return_drop=True)
print(dropped)
print(data_first_selected.shape)
```

```python
data_selected = data_first_selected.drop('day', axis=1)
test_selected = test4.drop('default', axis=1).drop('contact', axis=1).drop('day', axis=1)
test_selected.head()
```

## 2.2 Model Training

### 2.2.1 Split train and validation data

```python
train_svc, validation_svc = train_test_split(data_selected, test_size=0.2, random_state=7)
```

```python
print(f'train_svc.shape: {train_svc.shape}')
print(f'validation_svc.shape: {validation_svc.shape}')
```

### 2.2.2 Set X_train and Y_train¶

```python
x_train_svc = train_svc.drop('y', axis=1)
y_train_svc = train_svc['y']

x_val_svc = validation_svc.drop('y', axis=1)
y_val_svc = validation_svc['y']

x_test_svc = test_selected.drop('y', axis=1)
y_test_svc = test_selected['y']
```

### 2.2.3 Train Model

```python
for kernel in ["linear","poly","rbf","sigmoid"]:
    clf = SVC(kernel = kernel
    ,gamma="auto"
    ,degree = 1
    ,cache_size = 5000).fit(x_train_svc, y_train_svc)
    y_val_svc_pred = clf.predict(x_val_svc)
    score = clf.score(x_val_svc,y_val_svc)
    recall = recall_score(y_val_svc, y_val_svc_pred)
    f1 = f1_score(y_val_svc, y_val_svc_pred)
    auc = roc_auc_score(y_val_svc,clf.decision_function(x_val_svc))
    print("%s 's testing accuracy %f, recall is %f',f1 score is %f', auc is %f"% (kernel,score,recall,f1,auc))
```

```python
svc = SVC(kernel="rbf")
svc.fit(x_train_svc, y_train_svc)
y_train_rbf_pred = svc.predict(x_train_svc)
y_val_rbf_pred = svc.predict(x_val_svc)
```

### 2.2.3 Model Optimization

```python
inputs = data.columns[:-1]
outputs = data.columns[-1]
X_train = train[inputs]
y_train = train[outputs]
X_test = test[inputs]
y_test = test[outputs]
```

### 3.1.3 Independent variable normalization

```python
column_name=X_train.columns
zscore = preprocessing.StandardScaler()
X_train = zscore.fit_transform(X_train)
X_train = pd.DataFrame(data=X_train,columns=column_name)
X_test = zscore.fit_transform(X_test)
X_test = pd.DataFrame(data=X_test,columns=column_name)
```

```python
X_train.head()
```

## 3.2 Model Training
Here we use Decision Tree, Random Forest, AdaBoost and Gradient Boosting Decision Tree to train the model.

First, we train them with default parameters.

```python
methods = ['tree','Boosting','random forest','GBDT']
```

### 3.2.1 Fit model

```python
def machine_learning_default(method,x_train,y_train):
    if method == 'tree':
        clf = DecisionTreeClassifier()
        name = 'Decision Tree'
    elif method == 'Boosting':
        clf = AdaBoostClassifier()
        name = 'Adaboost'
    elif method == 'random forest':
        clf = RandomForestClassifier()
        name = 'RandomForest'
    elif method == 'GBDT':
        clf =GradientBoostingClassifier()
        name = 'GradientBoosting'
    else:
        print('Wrong Method!')
        return None
    clf.fit(x_train,y_train)
    return clf,name
```

```python
clfs_default = []
names = []
for method in methods:
    res = machine_learning_default(method,X_train,y_train)
    clfs_default.append(res[0])
    names.append(res[1])
```

### 3.2.2 Plot roc curve

```python
def auc_roc(clf,name):
    fpr,tpr,threshold = roc_curve(y_test,clf.predict_proba(X_test)[:,1])
    auc_value = auc(fpr,tpr)
    aucs.append(auc_value)
    plt.title('ROC for '+name)
    plt.plot(fpr,tpr,'b',label='AUC={}'.format(auc_value))
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.0])
    plt.xticks([])
    plt.yticks([])
    plt.ylabel('False Positive Rate')
    plt.xlabel('True Positive Rate')
    return
```

```python
aucs=[]
plt.figure()
plt.suptitle('ROC Curve')
for i in range(len(methods)):
    plt.subplot(2,2,i+1)
    auc_roc(clfs_default[i],names[i])
    plt.show()
```

```python
aucs=[]
plt.figure()
plt.suptitle('ROC Curve')
for i in range(len(methods)):
    plt.subplot(2,2,i+1)
    auc_roc(clfs_default[i],names[i])
    plt.show()
```

From the AUC value, Random Forest, Adaboost method and Gradient Boosting Decision Tree model have better predictive ability.

### 3.2.3 Plot Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
plt.figure()
plt.suptitle('Confussion Matrix')
for i in range(len(methods)):
    plt.subplot(2,2,i+1)
    cm=confusion_matrix(y_test,clfs_default[i].predict(X_test))
    ax=sns.heatmap(cm,annot=True,fmt='g',cmap='Blues')
    plt.title(names[i])
    plt.show()
```

From the confusion matrix, the AdaBoost method can best identify "no" examples, while the Random Forest method can best identify "yes" examples.

### 3.2.4 F1 score

```python
def model_evaluation(clf):
    pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test,pred)
    recall = recall_score(y_test,pred)
    precision = precision_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    return [accuracy,recall,precision,f1]
```

```python
res_default=pd.DataFrame(columns=['accuracy','recall','precision','f1_score'],index=names)
for i in range(len(methods)):
    res_default.loc[names[i]]=model_evaluation(clfs_default[i])
res_default['auc']=np.array(aucs)
res_default
```

In F1 score, the Random Forest method behaves better, the AdaBoost method has a low recall value due to its excessive tendency to predict "no", which affects the F1 score.

## 3.3 Model Selection

The following is a model selection for some parameters of the model. The training method is to select the model within the set parameter range in the form of 5-fold cross-validation on the training set and the auc value as the standard.

```python
clf_best=[]
```

### 3.3.1 Decision Tree
For the Decision Tree method, the criteria (Gini index/information gain), maximum depth (3, 5, 8, 15) and minimum leaf nodes (1, 2, 5, 10) are selected separately

```python
parameters = {'criterion':('gini','entropy'),'max_depth': [3,5,8,15],'min_samples_leaf':[1,2,5,10]}
clf_tree = DecisionTreeClassifier()
clf = GridSearchCV(clf_tree,parameters,cv=5,scoring='roc_auc')
clf.fit(X_train,y_train)
```

```python
print(clf.best_estimator_)
clf_best.append(clf.best_estimator_)
```

### 3.3.2 AdaBoost
For the AdaBoost method, we choose the number of boosts and the learning rate (speed of gradient convergence).

```python
parameters = {'n_estimators':[50, 80, 100, 200],'learning_rate':[0.2, 0.5, 0.8, 1]}
clf_boost = AdaBoostClassifier()
clf = GridSearchCV(clf_boost,parameters,cv=5,scoring='roc_auc')
clf.fit(X_train,y_train)
```

```python
print(clf.best_estimator_)
clf_best.append(clf.best_estimator_)
```

### 3.3.3 Random Forest

```python
parameters = {'n_estimators':[10,20,50,80,100],'criterion': ('gini','entropy'),'max_depth':[3,5,8,15],'min_samples_leaf':[1,2,5,10]}
clf_rf = RandomForestClassifier()
clf = GridSearchCV(clf_rf,parameters,cv=5,scoring='roc_auc')
clf.fit(X_train,y_train)
```

```python
print(clf.best_estimator_)
clf_best.append(clf.best_estimator_)
```

### 3.3.4 Gradient Boosting Decision Tree
For the GBDT method, we choose the same parameters as AdaBoost.

```python
parameters = {'n_estimators':[50, 80, 100, 200],'learning_rate':[0.2, 0.5, 0.8, 1]}
clf_boost = GradientBoostingClassifier()
clf = GridSearchCV(clf_boost,parameters,cv=5,scoring='roc_auc')
clf.fit(X_train,y_train)
```
