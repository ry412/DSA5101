## 1.3  Feature Engineering

### 1.3.1 Handling Missing Values

```python
 for col in data.columns:
        num_unknown = data[data[col]=='unknown'][col].count()
        num_1 = data[data[col]==-1][col].count()
        if num_unknown != 0:
            print(f'unknown value count in {col}: {num_unknown},in percentage of {num_unknown/len(data)}')
        if num_1 != 0:
            print(f'-1 value count in {col}: {num_1},in percentage of {num_1/len(data)}')
```

```python
 for col in test.columns:
        num_unknown_t = test[test[col]=='unknown'][col].count()
        num_1_t = test[test[col]==-1][col].count()
        if num_unknown_t != 0:
            print(f'unknown value count in {col}: {num_unknown_t},in percentage of {num_unknown_t/len(test)}')
        if num_1_t != 0:
            print(f'-1 value count in {col}: {num_1_t},in percentage of {num_1_t/len(test)}')
```

Here, we drop feature poutcome and pdays as it has over 70% meaningless value.

```python
data0 = data.drop(columns='poutcome').drop(columns='pdays')
test0 = test.drop(columns='poutcome').drop(columns='pdays')
```

```python
data0.shape
```

For job column has little unknown value, we delete the row containing 'unknown'.

```python
data1 = data0[~data0['job'].isin(["unknown"])]
test1 = test0[~test0['job'].isin(["unknown"])]
```

```python
data1.shape
```

For unknown values in conlumn educaton and contact, we are going to replace them with plural.

```python
data2 = data1.copy()
data2['education'].replace('unknown', data2["education"].value_counts().index[0],inplace=True)
data2['contact'].replace('unknown', data2["contact"].value_counts().index[0],inplace=True)
test2 = test1.copy()
test2['education'].replace('unknown', test2["education"].value_counts().index[0],inplace=True)
test2['contact'].replace('unknown', test2["contact"].value_counts().index[0],inplace=True)
```

### 1.3.2  Feature Encoding 

#### 1.3.2.1 Dependent Variable

```python
data2['y'] = data2['y'].map(lambda x: 1 if x == 'yes' else 0)
data2.head()
```

```python
test2['y'] = test2['y'].map(lambda x: 1 if x == 'yes' else 0)
test2.head()
```

### 1.3.3 Feature Binning

```python
c = toad.transform.Combiner()
c.fit(data2, y =  'y', method = 'dt',min_samples = 0.03, n_bins = 10)
data3 = c.transform(data2)
test3 = c.transform(test2)
data3.head()
```

```python
t = toad.transform.WOETransformer()
data_woe = t.fit_transform(data3,data3['y'], exclude='y')
test_woe = t.transform(test3)
test_woe.head()
```

### 1.3.4 Feature Selection

Here, we caluculate IV to select features. Similar to gini and entropy, IV reflects how well this feature can classify our target variable. The larger the IV, the stronger of a classifier the feature is.

```python
quality = toad.quality(data_woe, target = 'y')
quality
```

Here, we conduct preliminary feature selection according to IV and correlation (with other features), the variables are:
1. iv=0.02: the features with IV smaller than 0.02 are eliminated;
2. corr=0.7: if two or more features have Pearson correlation larger than 0.7, the ones with lower IV are eliminated;

```python
data_first_selected, dropped = toad.selection.select(data_woe,target = 'y', iv = 0.02, corr = 0.7, return_drop=True)
print(dropped)
print(data_first_selected.shape)
```

As the variable day represents last contact day of the month, it is meaningless for us to predict subscribe result. Thus, we also remove the variable day. 

```python
data_selected = data_first_selected.drop('day', axis=1)
test_selected = test_woe.drop('default', axis=1).drop('contact', axis=1).drop('day', axis=1)
data_selected.head()
```

# 2 Modeling

## 2.1 SVM

### Set X_train and Y_train

```python
x_train = data_selected.drop('y', axis=1)
x_test = test_selected.drop('y', axis=1)

y_train = data_selected['y']
y_test = test_selected['y']
```

### Train Model

```python
for kernel in ["linear","poly","rbf","sigmoid"]:
    clf = SVC(kernel = kernel
    ,gamma="auto"
    ,degree = 1
    ,cache_size = 5000).fit(x_train, y_train)
    y_hat_svc_train = clf.predict(x_train)
    y_hat_svc_test = clf.predict(x_test)
    score = clf.score(x_test,y_test)
    recall = recall_score(y_test, y_hat_svc_test)
    f1 = f1_score(y_test, y_hat_svc_test)
    auc = roc_auc_score(y_test,clf.decision_function(x_test))
    print("%s 's testing accuracy %f, recall is %f',f1 score is %f', auc is %f"% (kernel,score,recall,f1,auc))
```

```python
svc = SVC(kernel="sigmoid")
svc.fit(x_train, y_train)
y_hat_svc_sigmoid_train = svc.predict(x_train)
y_hat_svc_sigmoid_test = svc.predict(x_test)
```

### Evaluate Model

#### Training Set

```python
print(f'Precision for the training set of SVC: {precision_score(y_train, y_hat_svc_sigmoid_train)}')
print(f'Recall for the training set of SVC: {recall_score(y_train, y_hat_svc_sigmoid_train)}')
print(f'F-1 for the training set of SVC: {f1_score(y_train, y_hat_svc_sigmoid_train)}')
print(f'Accuracy for the training set of SVC: {accuracy_score(y_train,y_hat_svc_sigmoid_train)}')
print(f'Roc_Auc_score for the training set of SVC: {roc_auc_score(y_train,svc.decision_function(x_train))}')
```

#### Test Set

```python
print(f'Precision for the test set: {precision_score(y_test, y_hat_svc_sigmoid_test)}')
print(f'Recall for the test set: {recall_score(y_test, y_hat_svc_sigmoid_test)}')
print(f'F-1 for the test set: {f1_score(y_test,y_hat_svc_sigmoid_test)}')
print(f'Accuracy for the test set: {accuracy_score(y_test, y_hat_svc_sigmoid_test)}')
print(f'Roc_Auc_score  for the training set of SVC: {roc_auc_score(y_test,svc.decision_function(x_test))}')
```

### Model Optimization

```python
svc.get_params()
```

```python
C_range = np.linspace(0.01,1,10)
 
recallall = []
aucall = []
scoreall = []
for C in C_range:
    svc_c = SVC(kernel = "sigmoid",C=C,cache_size = 5000
    ,class_weight = "balanced"
    ).fit(x_train, y_train)
    y_hat_svc_sigmoid_test_c = svc_c.predict(x_test)
    score = svc_c.score(x_test,y_test)
    recall = recall_score(y_test, y_hat_svc_sigmoid_test_c)
    auc = roc_auc_score(y_test,clf.decision_function(x_test))
    recallall.append(recall)
    aucall.append(auc)
    scoreall.append(score)
    print("under C %f, testing accuracy is %f,recall is %f', auc is %f" % (C,score,recall,auc))
 
print(max(aucall),C_range[aucall.index(max(aucall))])
plt.figure()
plt.plot(C_range,recallall,c="red",label="recall")
plt.plot(C_range,aucall,c="black",label="auc")
plt.plot(C_range,scoreall,c="orange",label="accuracy")
plt.legend()
plt.show()
```

```python
svc_a = SVC(kernel="sigmoid",C=0.01,class_weight = "balanced")
svc_a.fit(x_train, y_train)
y_hat_svc_sigmoid_train_a = svc_a.predict(x_train)
y_hat_svc_sigmoid_test_a = svc_a.predict(x_test)
```

```python
print(f'Precision for the test set: {precision_score(y_test, y_hat_svc_sigmoid_test_a)}')
print(f'Recall for the test set: {recall_score(y_test, y_hat_svc_sigmoid_test_a)}')
print(f'F-1 for the test set: {f1_score(y_test,y_hat_svc_sigmoid_test_a)}')
print(f'Accuracy for the test set: {accuracy_score(y_test, y_hat_svc_sigmoid_test_a)}')
print(f'Roc_Auc_score  for the training set of SVC: {roc_auc_score(y_test,svc_a.decision_function(x_test))}')
```
